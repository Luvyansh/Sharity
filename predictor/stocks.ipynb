```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
sp500
```

```python
sp500.plot.line(y="Close", figsize=(10, 5), use_index=True)
```

```python
#Cleaning Data
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tommrow"] = sp500["Close"].shift(-1)
sp500["Target"] = (sp500["Tommrow"] > sp500["Close"]).astype("int")
sp500.dropna(inplace=True)
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC

model = RFC(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
import pandas as pd

pres = model.predict(test[predictors])
pres = pd.Series(pres, index=test.index)

pres
```

```python
precision_score(test["Target"], pres)
```

```python
combined = pd.concat([test["Target"], pres], axis=1)
combined.plot.line(figsize=(10, 5))
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    pres = model.predict(test[predictors])
    pres = pd.Series(pres, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], pres], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Close_Trend_{horizon}"
    sp500[trend_column] = sp500["Close"] - rolling_averages["Close"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
model = RFC(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    pres = model.predict_proba(test[predictors])[:, 1]
    pres[pres >= .6] = 1
    pres[pres < .6] = 0
    pres = pd.Series(pres, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], pres], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
